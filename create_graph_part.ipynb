```python
import networkx as nx
import matplotlib.pyplot as plt

# Example dictionary (adjacency list format)
graph_dict = {
    "A_table": ["B_table", "C_table"],
    "B_table": ["D_table"],
    "C_table": ["D_table", "E_table"],
    "D_table": [],
    "E_table": ["A_table"]
}

# Create a directed graph
G = nx.DiGraph()

# Add edges from dictionary
for node, neighbors in graph_dict.items():
    for neighbor in neighbors:
        G.add_edge(node, neighbor)

# Set figure size
plt.figure(figsize=(8, 6))

# Draw the graph with smaller labels
nx.draw(
    G,
    with_labels=True,
    node_size=2000,
    node_color="lightblue",
    arrows=True,
    font_size=8,
    verticalalignment="center",
    horizontalalignment="center"
)

plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Example dictionary (adjacency list format)
graph_dict = {
    "A_table": ["B_table", "C_table"],
    "B_table": ["D_table"],
    "C_table": ["D_table", "E_table"],
    "D_table": [],
    "E_table": ["A_table"]
}

# Create a directed graph
G = nx.DiGraph()
for node, neighbors in graph_dict.items():
    for neighbor in neighbors:
        G.add_edge(node, neighbor)

# Set figure size
plt.figure(figsize=(8, 6))

# Use layout
pos = nx.spring_layout(G)

# Draw edges only
nx.draw_networkx_edges(G, pos, arrows=True)

# Draw labels inside resizable rectangles
nx.draw_networkx_labels(
    G,
    pos,
    font_size=8,
    verticalalignment="center",
    horizontalalignment="center",
    bbox=dict(boxstyle="round,pad=0.3", fc="lightblue", ec="black")
)

plt.axis("off")
plt.show()
```
